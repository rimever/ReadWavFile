```python
file_path = "./input/youfit.wav"
wav_data = open(file_path, "rb")
contents = wav_data.read()
wav_data.close()
print(contents)
```

```python
print('RIFF = ' + repr(contents[0:4]))
print(' = ' + str(int.from_bytes(contents[4:8], byteorder='little')))
print(' = ' + repr(contents[8:12]))
print('fmt = ' + repr(contents[12:16]))
print('fmt = ' + str(int.from_bytes(contents[16:20], byteorder='little')))
print(' = ' + str(int.from_bytes(contents[20:22], byteorder='little')))
print(' = ' + str(int.from_bytes(contents[22:24], byteorder='little')))
print('(Hz) = ' + str(int.from_bytes(contents[24:28], byteorder='little') / 1000))
print('1 = ' + str(int.from_bytes(contents[28:32], byteorder='little')))
print(' = ' + str(int.from_bytes(contents[32:34], byteorder='little')))
print('/ = ' + str(int.from_bytes(contents[34:36], byteorder='little')))

print('2 = ' + repr(contents[36:40]))
print('2 = ' + str(int.from_bytes(contents[40:44], byteorder='little')))
```

```python
print('[s]' + str(int.from_bytes(contents[40:44], byteorder='little') / int.from_bytes(contents[28:32], byteorder='little')))
```
